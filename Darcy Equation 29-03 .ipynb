```python
def clamond(R,k,d):
    eD=k/d
    x1=eD*(R)* 0.123968186335417556;
    x2=(np.log(x1))-0.779397488455682028;

    f = x2 - 0.2;
    g = (np.log(x1+f)-0.2)/(1+x1+f);
    f = f-(1+x1+f+0.5*g)*g*(x1+f)/(1+x1+f+g*(1+g/3));
    g = (np.log(x1+f)-0.2)/(1+x1+f);
    f = f-(1+x1+f+0.5*g)*g*(x1+f)/(1+x1+f+g*(1+g/3));
    f = 1.151292546497022842/f;                   
    f = f*f; 
    return f
  
j=clamond(5600,0.01,1)
print(j)
```
